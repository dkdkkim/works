# Data  생성하기

```python
import numpy as np
import matplotlib.pyplot as plt
import math

### standard point
x = [0, 10]
y = [0, 4]

plt.plot(x,y,'ro')

### uniform rand
uni_angle = np.random.uniform(low=-0.1, high=1.1, size = 200) * (math.pi)
uni_r = np.random.uniform(low = 7, high = 12, size = 200)
uni_x = []
uni_y = []

for i in range(0,200):
    uni_x.append(uni_r[i] * math.cos(uni_angle[i]))
    uni_y.append(uni_r[i] * math.sin(uni_angle[i]))

plt.plot(uni_x, uni_y, 'bo')

### Gausian rand
gaus_angle = np.random.normal(loc = 0.5, scale = 0.15, size = 200)* math.pi + math.pi
gaus_r = np.random.normal(loc = 9.5, scale = 0.7, size = 200)
gaus_x = []
gaus_y = []

for i in range(0,200):
    gaus_x.append(10+gaus_r[i] * math.cos(gaus_angle[i]))
    gaus_y.append(4+gaus_r[i] * math.sin(gaus_angle[i]))

plt.plot(gaus_x,gaus_y,'go')
plt.show()

data = np.empty((400,2))
result = np.empty(400)

for i in range(0,200):
    data[i][0] = uni_x[i]
    data[i][1] = uni_y[i]
    data[i+200][0] = gaus_x[i]
    data[i+200][1] = gaus_y[i]
    result[i]=-1
    result[i+200]=1
    
data = data - data.mean()
data = data / np.abs(data).max()    

print result
```

# tanh func

```python
### Hyperbolic tangent
def hyper_tan(input,alpha):

    output = np.empty(input.shape)
    for i in range(0,input.shape[0]):
        for j in range(0,input.shape[1]):
            output[i][j] = (1-math.exp((-1)*alpha*input[i][j]))/(1+math.exp((-1)*alpha*input[i][j]))
            if output[i][j] >= 0: output[i][j] = 1
            else : output[i][j] = -1
#     if input[i][j] >= 0: output[i][j] = 1
#     else : output[i][j] = -1        

        
    
    return output
```

# training

```python
### weights initiation
import matplotlib.pyplot as plt

N_layer = 3
N_nodes = np.array([2,4])

weights_input = np.random.uniform(-1,1,(len(data[0]),N_nodes[0]))
weights_hidden = np.random.uniform(-1,1,(N_nodes[0],N_nodes[1]))
weights_output = np.random.uniform(-1,1,(N_nodes[1],1))

bias = np.random.randn(N_layer)
output = np.zeros((N_layer,np.max(N_nodes),len(data)))
epoch = 50
acc = []
loss = []
alpha = 1

for e in range(1,epoch):
    
    print 'epoch = ', e
    
    ### training
    
    ## input layer
    output = data.dot(weights_input)
#     output =  hyper_tan(output,alpha)

#     print output
#     print output.shape
    
    ## hidden layer
    output = output.dot(weights_hidden)
#     output = hyper_tan(output,alpha)
#     print output
#     print output.shape
    
    ## output layer
    output = output.dot(weights_output)
    output = hyper_tan(output,alpha)
#     print output
#     print output.shape
    

    ## accuracy cal
    acc_count = 0
    for i in range(0,len(data)):
        if result[i] == output[i]: acc_count += 1

    acc.append(float(acc_count)/float(len(data)))    
    
    ## loss cal
    error = np.empty(len(data))
    error_sum = 0
    
    for i in range(0,len(data)):
        error[i] = result[i]-output[i]
        error_sum = error_sum + 0.5*(error[i])**2
    
    loss.append(error_sum)

    
    ### optimizing
    
    lr = 0.001
    # input layer
    for n in range(0,len(data)):
        bias[0] = bias[0] + lr*(result[n]-output[n])*(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
        for k in range(0,len(data[0])):
            for j in range(0,N_nodes[0]):
                weights_sum = 0
                for i in range(0,N_nodes[1]):
                    weights_sum += weights_output[i]*weights_hidden[j][i]
                weights_input[k][j] = weights_input[k][j] + lr*(result[n]-output[n])*data[n][k]*weights_sum \
                *(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
                
    #hidden layer
    for n in range(0,len(data)):
        bias[1] = bias[1] + lr*(result[n]-output[n])*(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
        for j in range(0,N_nodes[0]):
            for i in range(0,N_nodes[1]):
                weights_sum = 0
                for k in range(0,len(data[0])):
                    weights_sum += weights_input[k][j]*data[n][k]
                weights_hidden[j][i] = weights_hidden[j][i] + lr*(result[n]-output[n])*weights_output[i]*weights_sum \
                *(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
    
    #output layer
    for n in range(0,len(data)):
        bias[2] = bias[2] + lr*(result[n]-output[n])*(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
        for i in range(0,N_nodes[1]):
            weights_sum_j= 0
            for j in range(0,N_nodes[0]):
                weights_sum_k = 0
                for k in range(0,len(data[0])):
                    weights_sum_k += weights_input[k][j]*data[n][k]
                weights_sum_j += weights_hidden[j][i]*weights_sum_k
            weights_output[i] =  weights_output[i] + lr*(result[n]-output[n])*weights_sum_j \
            *(1-np.tanh(alpha*output[n]))*(1+np.tanh(alpha*output[n]))*alpha
    
    
#     print acc
#     print loss
#     print weights_input
#     print weights_hidden
#     print weights_output
#     print bias

plt.figure(1)
plt.plot(acc)
plt.title('accuraccy')

plt.figure(2)
plt.plot(loss)
plt.title('loss')


```
