# Data  생성하기

```python
import numpy as np
import matplotlib.pyplot as plt
import math

### standard point
x = [0, 10]
y = [0, 4]

plt.plot(x,y,'ro')

### uniform rand
uni_angle = np.random.uniform(low=-0.1, high=1.1, size = 200) * (math.pi)
uni_r = np.random.uniform(low = 7, high = 12, size = 200)
uni_x = []
uni_y = []

for i in range(0,200):
    uni_x.append(uni_r[i] * math.cos(uni_angle[i]))
    uni_y.append(uni_r[i] * math.sin(uni_angle[i]))

plt.plot(uni_x, uni_y, 'bo')

### Gausian rand
gaus_angle = np.random.normal(loc = 0.5, scale = 0.15, size = 200)* math.pi + math.pi
gaus_r = np.random.normal(loc = 9.5, scale = 0.7, size = 200)
gaus_x = []
gaus_y = []

for i in range(0,200):
    gaus_x.append(10+gaus_r[i] * math.cos(gaus_angle[i]))
    gaus_y.append(4+gaus_r[i] * math.sin(gaus_angle[i]))

plt.plot(gaus_x,gaus_y,'go')
plt.show()

data = np.empty((400,2))
result = np.empty(400)

for i in range(0,200):
    data[i][0] = uni_x[i]
    data[i][1] = uni_y[i]
    data[i+200][0] = gaus_x[i]
    data[i+200][1] = gaus_y[i]
    result[i]=-1
    result[i+200]=1
    
data = data - data.mean()
data = data / np.abs(data).max()    
print data[399]
print result
print len(data), len(result)
```

```python
### weights initiation
import matplotlib.pyplot as plt

weights = np.random.normal(0,1,len(data[0]))
bias = np.random.normal(0,1,1)
epoch = 50
acc = []
loss = []

for k in range(0,epoch):
    
    ### forward
    output =np.empty(len(data))

    for i in range(0,len(data)):
        output[i] = data[i][0]*weights[0]+data[i][1]*weights[1] + bias

    ### Sigmoid
    alpha = 1

    for i in range(0,len(data)):

        output[i] = np.tanh(output[i])

        if output[i] >= 0: output[i] = 1
        else : output[i] = -1
    
    ## accuracy cal
    acc_count = 0
    for i in range(0,len(data)):
        if result[i] == output[i]: acc_count += 1

    acc.append(float(acc_count)/float(len(data)))    
    
    ## Backpropagation
    error = np.empty(len(data))

    ## loss cal
    error_sum = 0
    
    for i in range(0,len(data)):
        error[i] = result[i]-output[i]
        error_sum = error_sum + 0.5*(error[i])**2
    
    loss.append(error_sum)

    
    ### optimizing
    lr = 0.05
    for i in range(0,len(data)):
        bias = bias + lr*(result[i]-output[i])*(1-np.tanh(output[i]))*(1+np.tanh(output[i]))*alpha
        for j in range(0,len(weights)):
            weights[j] = weights[j] + lr*(result[i]-output[i])*data[i][j]*(1-np.tanh(alpha*output[i]))*(1+np.tanh(alpha*output[i]))*alpha

plt.figure(1)
plt.plot(acc)
plt.title('accuraccy')

plt.figure(2)
plt.plot(loss)
plt.title('loss')

print acc
print loss
```
