# Making data

```python
import numpy as np
import matplotlib.pyplot as plt
import math

### standard point
x = [0, 10]
y = [0, 4]

plt.plot(x,y,'ro')

### uniform rand
uni_angle = np.random.uniform(low=-0, high=1, size = 200) * (math.pi)
uni_r = np.random.uniform(low = 7, high = 12, size = 200)
uni_x = []
uni_y = []

for i in range(0,200):
    uni_x.append(uni_r[i] * math.cos(uni_angle[i]))
    uni_y.append(uni_r[i] * math.sin(uni_angle[i]))

plt.plot(uni_x, uni_y, 'bo')

### Gausian rand
# gaus_angle = np.random.normal(loc = 0.5, scale = 0.15, size = 200)* math.pi + math.pi
# gaus_r = np.random.normal(loc = 9.5, scale = 0.7, size = 200)
gaus_angle = np.random.uniform(low=-0, high=1, size = 200) * (math.pi) + math.pi
gaus_r = np.random.uniform(low = 7, high = 12, size = 200)
gaus_x = []
gaus_y = []

for i in range(0,200):
    gaus_x.append(10+gaus_r[i] * math.cos(gaus_angle[i]))
    gaus_y.append(4+gaus_r[i] * math.sin(gaus_angle[i]))

plt.plot(gaus_x,gaus_y,'go')
plt.show()

data_o = np.empty((400,2))
result = np.empty(400)

for i in range(0,200):
    data_o[i][0] = uni_x[i]
    data_o[i][1] = uni_y[i]
    data_o[i+200][0] = gaus_x[i]
    data_o[i+200][1] = gaus_y[i]
    result[i] = -1
    result[i+200] = 1

mean = data_o.mean()
data = data_o# - mean
maax = np.abs(data).max()
data = data# / maax    
# print data
print result
plt.figure(2)
plt.plot(data[:,0],data[:,1],'ro')
```

# training

```python
### weights initiation
import matplotlib.pyplot as plt

N_input = 2
N_nodes = 10
N_class = 1

# weights_input = np.random.uniform(-1,1,(data.shape[1],N_nodes[0]))
# weights_hidden = np.random.uniform(-1,1,(N_nodes[0],N_nodes[1]))
# weights_output = np.random.uniform(-1,1,(N_nodes[1],1))
# bias = np.random.randn(N_layer)

W0 = np.random.uniform(-1,1,(N_input,N_nodes))
W1 = np.random.uniform(-1,1,(N_nodes,N_class))

b0 = np.ones((1,N_nodes))*0.00
b1 = np.ones((1,N_class))*0.00

epoch = 1000
acc = []
loss = []
alpha = 5
lr = 0.01



for e in range(1,epoch):
    
    if e % 100 == 0:
        lr *= 0.5
        print "learning_rate :", lr    
        print "epoch :", e
        print acc[-1]
        print loss[-1]
        
        
    acc_count = 0
    loss_count = 0
    
    arr = np.arange(len(data))
    np.random.shuffle(arr)
    
    for n in arr:
        # Do the first Linear step 
        z0 = data[n].dot(W0) + b0
        a0 = np.tanh(z0)

        # Second linear step
        z1 = a0.dot(W1) + b1
        output = np.tanh(z1)
        
        # Calculate loss derivative with respect to output
        dL = result[n] - output[0]
        
        if abs(dL) < 1 : acc_count += 1
        loss_count += 0.5*((dL[0])**2)    
        dz1 = alpha*dL*(1-output[0]**2)        
        # Calculate loss derivative with respect to second layer weights
        dW1 = dz1*a0.T #dW2 = 1/m*(a1.T).dot(dz2) 
        # Calculate loss derivative with respect to second layer bias
        db1 = dz1

        # Calculate loss derivative with respect to first layer
        dz0 = alpha*dz1[0]*np.multiply((1-a0**2),W1.T)
#         dz2 = np.multiply(dz3.dot(W3.T) ,1-a2**2)
        # Calculate loss derivative with respect to first layer weights
        dW0 = np.dot(np.array([data[n]]).T, dz0)
        # Calculate loss derivative with respect to first layer bias
        db0 = dz0
#         print data[n], W0, b0, z0, a0, W0 + lr*dW0
#         raw_input("enter")
        W1 += lr*dW1
        W0 += lr*dW0
        b1 += lr*db1
        b0 += lr*db0



   
    acc.append(float(acc_count)/float(len(data)))
    loss.append(loss_count)
    #     print output

plt.figure(1)
plt.plot(acc)
plt.title('accuraccy')

plt.figure(2)
plt.plot(loss)
plt.title('loss')

print acc[-1]
print loss[-1]
```
